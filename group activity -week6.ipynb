```python
#importing pandas,numply,matplotlib.pyplot,seaborn libaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading the dataset to the python environment
data=pd.read_excel('Dataset_w6.xlsx')
```

```python
#Displaying the columns in the dataset
data.columns
```

```python
#displaying  top 5  rows of the dataset
data.head()
```

```python
#To know data types of each columns and checking for null values in the columns
data.info()
```

```python
#renaming columns
data=data.rename(columns={'Population in thousands (2017)':'Population','GDP per capita (current US$)':'GDP','International trade: Imports (million US$)':'International_trade','Surface area (km2)':'Surface_area','Agricultural production index ':'Agricultural_production_index'})
```

```python
data.head()
```

**01. Find out how many countries are there in the dataset**

```python
num_countries=data.country.nunique()
print('There are {} countries in the dataset'.format(num_countries))
```

**02. Examine the data by using visualizations.**

```python
#Dataframe showing top 10 countries with highest GDP.
high_GDP_per_capita = data.sort_values('GDP', ascending=False).head(10)
```

```python
# Barplot
plt.title("Top 10 Countries with Hghest GDP ($ per capita)")
plt.xlabel("Country")
plt.ylabel("GDP ($ per capita)")
plt.xticks(rotation=75)
sns.barplot(x = high_GDP_per_capita.country, y = high_GDP_per_capita['GDP']);
```

```python
#Dataframe showing  10 countries with lowest GDP
lowest_GDP_per_capita = data.sort_values('GDP').head(10)
```

```python
# Barplot
plt.title(" 10 Countries with Lowest GDP ($ per capita)")
plt.xlabel("Country")
plt.ylabel("GDP ($ per capita)")
plt.xticks(rotation=75)
sns.barplot(x = lowest_GDP_per_capita.country, y = lowest_GDP_per_capita['GDP']);
```

```python
#Dataframe showing top 10 countries with highest population.
high_population = data.sort_values('Population', ascending=False).head(10)
```

Pie chart showing top 10 countries with highest population

```python
# Pie chart showing top 10 countries with highest population
plt.pie(high_population.Population, 
        labels=high_population.country, 
        autopct='%1.1f%%', 
        startangle=90, 
        radius=2.5);

plt.show()
```

```python
#Dataframe showing how many countries each region are having in our data.
countries_per_region = data.Region.value_counts()
```

```python
# Barplot
plt.figure(figsize=(10,7))
plt.title('Number of Countries by Region')
plt.ylabel('Region')
plt.xlabel('Number of countries')
sns.barplot(x= countries_per_region.values, y= countries_per_region.index);
```

```python
#scatter plot of Agricultural production index vs GDP
plt.figure(figsize=(14,10))
sns.scatterplot(data=data,x='Agricultural_production_index',y='GDP',hue='Region')
plt.title('Agricultural production index vs GDP')
plt.show()
```

```python
#how population distributes over regions in swarmplot

plt.figure(figsize = (10, 5))
sns.swarmplot(x = "Region", y = "Population", data = data)
plt.xticks(rotation = 90)
plt.xlabel("Regions")
plt.ylabel("Population")
plt.title("Population distribution over regions")
plt.show()
```

```python
#BAE GRAPH OF REGIONAL AVERAGE GDP PER CAPITA
fig = plt.figure(figsize=(12, 4))
data.groupby('Region')['GDP'].mean().sort_values().plot(kind='bar')
plt.title('Regional Average GDP per Capita')
plt.xlabel("Region")
plt.ylabel('Average GDP per Capita')
plt.show()
```

```python
#pairplot of dataset
sns.pairplot(data)
plt.tight_layout()
plt.show()
```

**05.Find out which region has more consistent with respect to GDP per capita (current US$)**

```python
CV=(data.groupby('Region')['GDP'].std())/(data.groupby('Region')['GDP'].mean())
```

```python
coeff_var=CV*100
coeff_var.sort_values()
```

Find-

*Oceania  region is  more consistent with respect to GDP per capita*
